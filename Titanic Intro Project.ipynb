```python
def male_female_child(passenger):
    age,sex=passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
sns.factorplot('Pclass',kind='count',data=titanic_df,hue='person',order=[1,2,3])
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])
    
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
sns.factorplot(x='Cabin',kind='count',data=cabin_df,palette='winter_d')
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']

cabin_ord = Series(cabin_df.values.ravel()).unique()
cabin_ord.sort()

sns.factorplot(x='Cabin',kind='count',data=cabin_df,palette='summer', x_order=cabin_ord)
```

2.) Did having a family member increase the odds of surviving the crash?

```python
titanic_df['Sibling'].loc[titanic_df['SibSp']>0] = 'Sib'
titanic_df['Parent'].loc[titanic_df['Parch']>0] = 'Par'
titanic_df['Sibling'].loc[titanic_df['SibSp']==0] = ''
titanic_df['Parent'].loc[titanic_df['Parch']==0] = ''

titanic_df['Family'] = titanic_df['Sibling']

titanic_df['Family'].loc[(titanic_df['Family']!='') & (titanic_df['Parent']!='')] = titanic_df['Family'] + ' + ' + titanic_df['Parent']
titanic_df['Family'].loc[titanic_df['Family']==''] = titanic_df['Parent']
titanic_df['Family'].loc[titanic_df['Family']==''] = 'Alone'

```

```python
sns.factorplot('Family','Survived',data=titanic_df)
```
